# IEA ETP 2020のウェブサイトから技術のリストを取得する

```python
import re
import json
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.iea.org/articles/etp-clean-energy-technology-guide"
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
```

```python
trl = soup.find_all("script")[5].string

trl = trl.replace('\nconst items = ','')
trl = re.sub(';\n$','',trl)
```

```python
with open('trl.json','w') as f:
    print(trl,file=f)
```
